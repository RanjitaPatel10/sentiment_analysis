```python
# find all the punctuation
punctuation = string.punctuation
print(punctuation)
print("-"* 100)
# find all the stopwords
stopwords = nltk.corpus.stopwords.words('english')
print(stopwords)
print("-"* 100)

# find all the lemmatizer
lm = WordNetLemmatizer()
print(lm)
print("-"* 100)
```

```python
text = df['review'][1]
print(text)
print("-"* 100)
soup = BeautifulSoup(text, 'html.parser').get_text()
print(soup)
```

```python
def clean_text(text):
    # remove html tags
    only_text = BeautifulSoup(text, 'html.parser').get_text()

    # remove punctuation
    text = "".join([char.lower() for char in only_text if char not in punctuation])

    # tokenizing
    tokens = nltk.word_tokenize(text)

    # remove stopwords and apply lamitization
    final_text = " ".join([lm.lemmatize(word) for word in tokens if word not in stopwords])
    return final_text
# clean_text(text)
```

```python
X = df['review'].apply(lambda x: clean_text(x))
X
```

```python
Y = df = df['sentiment'].apply(lambda x: 1 if x == 'positive' else 0)
Y
```

```python
##Average num of words in text
sum=0
for i in X.to_list():
    sum+=len(i.split(" "))
avg = sum/len(X.to_list())
avg
```
